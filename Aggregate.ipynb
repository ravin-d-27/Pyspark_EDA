# Aggregate Functions

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Aggregate').getOrCreate()
```

```python
df_pyspark = spark.read.csv('Food demand.csv',header=True,inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.groupBy('emailer_for_promotion').sum().show()
```

```python
df_pyspark.groupBy('emailer_for_promotion').mean().show()
```

```python
df_pyspark.groupBy('emailer_for_promotion').min().show()
```

```python
df_pyspark.groupBy('emailer_for_promotion').max().show()
```

```python
df_pyspark.groupBy('emailer_for_promotion').count().show()
```

```python
df_pyspark.agg({'checkout_price':'max'}).show()
```
