# Importing the Libraries

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Filters").getOrCreate()
```

```python
df_pyspark = spark.read.csv("Food Demand.csv", header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

### Filter Operations

```python
# Base price of the food which is greater than 500 

df_pyspark.filter("base_price>=500.0").show()
```

```python
# Selecting only the week and num_orders

df_pyspark.filter("base_price>=500.0").select(['week','num_orders']).show()
```

Another method

```python
df_pyspark.filter((df_pyspark['base_price']>=500)&(df_pyspark['base_price']<=600)).select(['checkout_price','base_price']).show()
```

```python
# Which are not in the range 500 to 600

df_pyspark.filter(~((df_pyspark['base_price']>=500)&(df_pyspark['base_price']<=600))).select(['checkout_price','base_price']).show()
```
